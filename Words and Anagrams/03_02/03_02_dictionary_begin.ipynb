```python
word = open('words','r')
word
```

```python
#reading all the lines
wordlist = word.readlines()
```

```python
wordlist
```

```python
len(wordlist)
```

```python
#removing /n from the wordlist and making all the characters to the lowercase
cleanword = [word.strip().lower() for word in wordlist]
```

```python
cleanword
```

```python
#removing duplicates
wordunique = list(set(cleanword))
wordunique
```

```python
wordunique.sort()
wordunique[:10]
```

```python
#magic of comprehensions, all the above comprehension in a single line
wordcleannew = sorted(list(set([word.strip().lower() for word in open('words','r')])))
wordcleannew[:10]
```
