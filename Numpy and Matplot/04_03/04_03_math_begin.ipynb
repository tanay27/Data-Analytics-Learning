```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#instruction to keep 
%matplotlib inline
```

```python
a = np.linspace(0,10,40)
```

```python
a
```

```python
#applying a simple trigonometry 
sina = np.sin(a)
sina
```

```python
cosa = np.cos(a)
```

```python
#plotting
plotting_sina = plt.plot(a,sina)
```

```python
plotting_cosa = plt.plot(a,np.cos(a))
```

```python
plt.plot(a,cosa,'x')
plt.plot(a,sina,'o')
```

```python
y = sina + cosa
z = cosa**2-sina**2

plt.plot(a,y)
plt.plot(a,z)

plt.legend(['sina+cosa', 'cosa**2-sina**2'])
```

```python
np.dot(sina,cosa)
```

```python
#outer product--every possible combination
mat_outer = np.outer(sina,cosa)
```

```python
mat_outer
```

```python
#broadcating
v = np.linspace(0,10,5)
v+1 #this will add 1 to every other no.
```

```python
#adding one dimensional array to a two dimensional array
vv = np.outer(v,v)
print(vv)
vv+v
```

```python
#to add it to a new column
vv + v[:np.newaxis]
```
